# Análisis Exploratorio de Datos - EDA

##### Nuestro objetivo es que este EDA nos ayude a obtenerlo datos que nos permitan elaborar un modelo que prediga el precio de un alquiler turístico en Madrid.

## 1. Carga de Datos

```python
#Cargamos librerías
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Cargar desde el .CSV a un DF de Pandas
df = pd.read_csv('air-bnb-listings.csv', encoding='utf-8', delimiter=";")

df.head() #Ver Cabecera
```

```python
# Información Inicial
df.info()
```

##### Nuestro DataFrame tiene 17 columnas y 1.414.018 filas. A simple vista podemos observar que algunas columnas tienen valores nulos o faltantes (la 1, 8, 9 y 12). Más adelante decidiremos cómo vamos a tratarlos. También revisamos el tipo de datos que manera cada columna y la cantidad de memoria que utiliza el DF.

##### Antes de llegar al punto de seleccionar columnas para retirar las que no nos interesan hemos visto oportuno limpiar la cabecera del .CSV para que los nombres de las columnas que vamos a analizar tengan el mismo formato y sean más manejables a la hora de escribirlas.

```python
# Normalización de Columnas con función
def normalize_column_name(column_name):
    # Pasamos a minúsculas
    column_name = column_name.lower()
    # Quitamos signos de puntuación
    column_name = ''.join(char for char in column_name if char.isalnum() or char == ' ')
    # Quitamos dobles espacios
    column_name = ' '.join(column_name.split())
    return column_name

# Invocamos la función
df.columns = [normalize_column_name(column) for column in df.columns]

print(df.columns)
```

```python
# Chequeamos que el cambio se haya producido en nuestro DF
df.head(2)
```

```python
# Seleccionar Columnas que nos quedamos para el Análisis (ahora que tienen títulos más manejables)
df_selected = df.loc[:, ['location', 'coordinates', 'name', 'room type', 'room price', 'number of reviews', 'minimum nights', 'availibility', 'rooms rent by the host']]

df_selected.head(2)
```

## 2. Tratamiento de los Datos

```python
# Limpieza de filas que no nos sirven 
# 1º Nos quedaremos solo con las filas que contengan 'Madrid' en la columna 'location'
df_madrid = df_selected[df_selected['location'].str.contains('Madrid', case=False)]

df_madrid.head()
```

```python
# Comprobamos cómo ha afectado el filtrado al DF
df_madrid.info()
```

##### Nuestro nuevo DF tiene un número de filas y columnas más reducido, así que ahora ocupa menos. La cabecera del DF ahora empieza en la fila 11.347 en lugar de en la 0. También nos encontramos con que por ahora solo manejaremos dos tipos de datos y con que solo la columna 'name' tiene valores NaN.

```python
# Veamos qué datos nos faltan y si pueden dar probelmas en el DF
# Mostrar nº de datos faltantes
missing_values = df_madrid.isnull().sum()
print("Valores nulos por columna:")
print(27 * '*')
print(missing_values)

# Mostrar las filas con valores faltantes
missing_name = df_madrid[df_madrid['name'].isnull()]
print("Filas con valores faltantes:")
print(27 * '*')
print(missing_name)
```

##### Como los datos faltantes son cadenas de texto, en principio no debe de haber problemas, porque no afectarían a las operaciones matemáticas (como sí sucedería con enteros o flotantes), aún así, para que no queden vacíos los rellenaremos con un nombre genérico.

```python
# Rellenar NaN con texto
df_madrid['name'].fillna('Nombre no disponible', inplace=True)

# Comprobemos que se ha rellenado en el DF llamando a los índices
indices = [302251, 1114717, 1310941]

# Comprobar columna 'name' para los índices dados
for idx in indices:
    print(f"Índice {idx}: {df_madrid.loc[idx, 'name']}")
    
# Mostrar de nuevo las filas
new_name = df_madrid.loc[[302251, 1114717, 1310941]]
print("Filas con nuevo nombre:")
print(27 * '*')
print(new_name)
```

```python
# Tratar la columna 'location' para quedarnos solo con el barrio
# Ya que solo trabajaremos con alojamientos en Madrid
df_madrid['location'] = df_madrid['location'].str.split(',').str[2]

df_madrid.head()
```

```python
# Comprobamos de nuevo que no hayamos perdido datos
df_madrid.info()
```

##### Seguimos contando con 21.255 filas, así que este último tratamiento no ha supuesto pérdida de información. Todas las filas contenían el barrio. 

```python
# Comprobemos cúales son los barrios, para terminar de confirmar que todo está correcto
conteo_location = df_madrid['location'].value_counts()

conteo_location
```

##### Todo correcto, pero ya hemos encontrado que hay barrios con poca información que podrían afectar al modelo. Los trataremos más adelante.

##### Llegando a este punto nos planteamos, ¿cuántas de las filas de nuestro DF tienen valor 0? Creemos que es importante comprobarlo para intentar comprender si esto puede afectar al modelo que intentamos construir.

```python
# Tratando los 0 que pueden afectar a nuestro modelado de datos
# Empezamos buscando enteros con valor = 0
values_0 = (df_madrid == 0)

# Columnas con 0 = true
conteo_true = values_0.any()

for columna, tiene_true in conteo_true.items():
    if tiene_true:
        num_true = values_0[columna].sum()
        print(f"Columna '{columna}' tiene {num_true} valor/es 0")
```

##### Ahora que sabemos qué columnas contien ceros, podemos sacar conclusiones. En las columnas de disponibilidad y número de reseñas, el valor 0 no debe afectarnos al modelo, puesto que 0 indica que no está disponible e alojamiento y un alojamiento puede no tener reseñas. El 0 que sí es importante que sustituyamos es el de la columna precio. Un solo 0 no va afectar mucho a los cálculos y resultados que estamos llevando a cabo, pero lo más correcto es referenciarlo y tratarlo, ya que si estuviésemos ante un DF con un número elevado de ceros en la columna precio, estaríamos haciendo cálculos sesgados que inducirían a errores en nuestro modelo.

##### Como un Alojamiento Turístico no puede ser aqluilado gratis a precio 0, vamos a imputarle a esta fila la mediana de los precios. Preferimos usar la mediana en lugar de la media porque es menos sensible a valores atípicos. Sabemos que el rango de precios es muy amplio, así que la mejor opción sin duda es la mediana.

```python
# Calcular mediana
mediana_room_price = df_madrid['room price'].median()

# Imputar valores igual a 0 con la mediana
df_madrid.loc[df_madrid['room price'] == 0, 'room price'] = mediana_room_price

# Comprobar que ya no hay 0 en 'room price'
values_0 = (df_madrid == 0)
conteo_true = values_0.any()

for columna, tiene_true in conteo_true.items():
    if tiene_true:
        num_true = values_0[columna].sum()
        print(f"Columna '{columna}' tiene {num_true} valor/es 0")
```

## 3. Visualización Gráfica de Datos, buscando outliers y relaciones

##### Como posteriormente vamos a construir un modelo que prediga el precio, vamos a prestarle bastante atención a esta variable. 

```python
# Veamos lo más impoirtante del precio
df_madrid['room price'].describe()
```

##### Vemos que tenemos 21.255 registros, y que los precios fluctuan entre 8 y 9.999 euros, una fluctuacion muy grande. En esta tabla leemos que:

- mean: La media (promedio) de los valores en la columna "room price". En promedio, el precio de las habitaciones es de aproximadamente 163.99.
- std: La desviación estándar de los valores en la columna "room price". Esto mide la dispersión de los valores con respecto a la media. En este caso, la desviación estándar es de aproximadamente 559.47.
- min: El valor mínimo en la columna "room price". El precio mínimo de una habitación es de 8.
- 25%: El percentil 25, también conocido como el primer cuartil o el cuartil inferior. Esto indica que el 25% de los precios de las habitaciones son iguales o inferiores a 36.
- 50%: El percentil 50, que es la mediana. Esto indica que el 50% de los precios de las habitaciones son iguales o inferiores a 60.
- 75%: El percentil 75, también conocido como el tercer cuartil o el cuartil superior. Esto indica que el 75% de los precios de las habitaciones son iguales o inferiores a 100.
- max: El valor máximo en la columna "room price". El precio máximo de una habitación es de 9999.

##### Empezaremos por localizar y eliminar los outliers o valores atípicos en el precio.

```python
# Frecuencia de los precios
plt.figure(figsize=(12, 6))
sns.histplot(df_madrid['room price'], bins=50, color='skyblue', kde=True)
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.title('Distribución del Precio en Madrid')
plt.xticks(ticks=range(0, 10001, 1000))

plt.show()
```

```python
# Bloxplot solo del precio para ver outliers
plt.figure(figsize=(20, 5))
sns.boxplot(x=df_madrid["room price"], color='skyblue')
plt.title('Boxplot de Precios')
plt.xlabel('Precio')

plt.show()
```

##### Vemos que hay muchos valores fuera de rango, tantos que la visualización de estos resulta compleja, así que emplearemos el método del rango intercuartílico para eliminarlos. Para ello calcularemos el rango intercuartílico (IQR), que es la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1). Los outliers se definen como aquellos valores que están por encima de Q3 + 1.5 * IQR o por debajo de Q1 - 1.5 * IQR.

```python
# Eliminamos outliers con método dek IQR
Q1 = df_madrid['room price'].quantile(0.25)
Q3 = df_madrid['room price'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

df_madrid_clean = df_madrid[(df_madrid['room price'] >= lower_limit) & (df_madrid['room price'] <= upper_limit)]

# Visualizamos cantidad de datos en el nuevo DF
df_madrid_clean.info()
```

```python
# Bloxplot de precios sin valores atípicos
plt.figure(figsize=(20, 5))
sns.boxplot(x=df_madrid_clean["room price"], color='skyblue')
plt.title('Boxplot de Precios')
plt.xlabel('Precio')

plt.show()
```

##### Ahora vamos a intentar comprender cómo afecta cada tipo de variable al precio del alquiler turístico. Usaremos la matriz de correlación , que es una tabla que muestra las correlaciones entre todas las variables en un conjunto de datos. Cada celda en la tabla representa el coeficiente de correlación entre dos variables. Cuanto más alto sea el número más relacionadas están las variables. nos fijaremos cuáles afectan más al precio. Pero antes, necesitamos que las variables 'location' que contiene los barrios y 'room type', tenga un valor numérico para poder correlacionarla con el precio.

```python
# Convirtamos 'location' en números para una nueva coluna en otro DF (para no sobreescribir el bueno)
df_expanded = df_madrid_clean.explode('location')
unique_locations = df_expanded['location'].unique()
location_to_number = {location: i for i, location in enumerate(unique_locations)}

df_expanded['location_number'] = df_expanded['location'].map(location_to_number)
correlation = df_expanded['room price'].corr(df_expanded['location_number'])

print("Correlación entre ubicación y precio:", correlation)
```

```python
# Convirtamos 'room type' en números para una nueva coluna en otro DF (para no sobreescribir el anterior)
df_expanded_final = df_expanded['room type'].unique()
room_type_to_number = {room_type: i for i, room_type in enumerate(df_expanded_final)}

# Creamos una nueva columna en el DataFrame que mapea cada tipo de habitación a su número correspondiente
df_expanded['room_type_number'] = df_expanded['room type'].map(room_type_to_number)

# Calculamos la correlación entre el precio de la habitación y el tipo de habitación
correlation = df_expanded['room price'].corr(df_expanded['room_type_number'])

print("Correlación entre tipo de habitación y precio:", correlation)
```

```python
# Matriz de correlación
columns = ['location_number', 'room_type_number', 'room price', 'number of reviews', 'minimum nights', 'availibility', 'rooms rent by the host']
correlation_matrix = df_expanded[columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis')
plt.title('Mapa de Calor')

plt.show()
```

##### Con esta gráfica ya podemos obtener conclusiones. Lo que más influye en el precio es el 'room type' (tipo de habitación), seguido bastante de lejos por 'rooms rent by the host' (habitaciones alquiladas por el anfitrión) y 'availibility' (disponibilidad).

##### Visualicemos los datos por tipo de habitación.

```python
# Gráfico de barras de los tipos de habitación
plt.figure(figsize=(10, 5))
sns.countplot(data=df_madrid, y='room type', order=df_madrid['room type'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Recuento')
plt.ylabel('Tipo')
plt.title('Distribución de Apartamentos por Tipo')

plt.show()
```

```python
# Relación entre Tipo de Habitación y Precio de la Habitación
plt.figure(figsize=(12, 6))
df_madrid_clean.boxplot(column='room price', by='room type', rot=45)
plt.xlabel('Tipo de Habitación')
plt.ylabel('Precio de la Habitación')
plt.title('Relación entre Tipo de Habitación y Precio de la Habitación')
plt.suptitle('')
plt.show()
```

##### Lo que más influye en el precio es si se trata de un apartamento entero o una habitación privada, ya que habitación compartida y habitación de hotel no tienen apenas representación en el DataFrame. Así que debemos eliminarlos para simplificar el modelo de cara a que tenga una mayor precisión. Así que empezaremos a filtar los datos.

```python
# Limpiamos tipos de apartamentos que no aportan
df_madrid_filtered = df_madrid_clean[~df_madrid_clean['room type'].isin(['Shared room', 'Hotel room'])]

df_madrid_filtered
```

```python
# Analizando 'rooms rent by the host' y su relación con el precio
plt.figure(figsize=(10, 10))
sns.countplot(data=df_madrid_filtered, y='rooms rent by the host', order=df_madrid_filtered['rooms rent by the host'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Recuento')
plt.ylabel('Nº Habitaciones')
plt.title('Habitaciones que Alquila el Anfitrión')

plt.show()

plt.figure(figsize=(12, 6))
df_madrid_filtered.boxplot(column='room price', by='rooms rent by the host', rot=45)
plt.xticks(rotation=90)
plt.xlabel('Nº de Habitaciones')
plt.ylabel('Precio de la Habitación')
plt.title('Relación entre Nº de Habitaciones y Precio de la Habitación')
plt.suptitle('')

plt.show()

# Nº de habitaicones
room_rent_counts = df_madrid_filtered['rooms rent by the host'].value_counts()
print(room_rent_counts)
```

##### El número de habitaciones afecta ya que cuantas más habitaciones tiene el alojamiento más caro es su precio, seguramente esté también relacionado con el tamaño de las intalaciones y los m^2 totales que se están alquilando. Por eso también vamos a simplificar un poco. Aunque no dudamos de que puedan existir propiedades con 244 habitaciones, estamos casi seguras de que seguramente se trate de un error. Así que visualizando los datos, para nuestro modelo vamos a quedarnos con los datos de precio para los que tenemos más de 300 muestras. Esto supone cortar de DF qudándonos únicamente con las filas que se encuentren entre 1 a 7 habitaciones.

```python
# Filtamos para quedarnos con los nº de habitaciones más habituales
df_madrid_filtered = df_madrid_filtered[df_madrid_filtered['rooms rent by the host'] <= 7]

df_madrid_filtered
```

```python
# Analizando 'availibility' y su relación con el precio
plt.figure(figsize=(60, 8))
df_madrid_filtered.boxplot(column='room price', by='availibility', rot=90)
plt.xlabel('Disponibilidad')
plt.ylabel('Precio de la Habitación')
plt.title('Relación entre Disponibilidad y Precio de la Habitación')
plt.suptitle('')
plt.xticks(rotation=45)
plt.xticks(range(len(df_madrid_filtered['availibility'].unique())), df_madrid_filtered['availibility'].unique())

plt.show()
```

##### Aunque el gráfico no se distingue muy bien, a pesar de nuestros esfuerzos por hacer el eje X más largo y poder separar más las líneas, teniendo en cuenta que buscamos reducir la complejidad de los datos y su distribución, vamos a eliminar todas las disponibilidades que tengan una prevalencia menor a 300.

```python
# Calcula el recuento de cada valor único en la columna 'availability'
availibility_counts = df_madrid_filtered['availibility'].value_counts()

# Filtra el DataFrame para quedarte solo con los valores que se repiten al menos 300 veces
valores_repetidos = availibility_counts[availibility_counts >= 300].index
df_madrid_filtered = df_madrid_filtered[df_madrid_filtered['availibility'].isin(valores_repetidos)]

df_madrid_filtered
```

```python
# Disponibilidad, ahora solo tenemos en cuenta las más habituales
room_rent_availibility = df_madrid_filtered['availibility'].value_counts()
print(room_rent_availibility)
```
